##1. BERT Next sentence prediction

```python
!pip install transformers
```

```python
from transformers import BertTokenizer, BertForNextSentencePrediction
import torch

# Load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertForNextSentencePrediction.from_pretrained('bert-base-uncased')

# Example sentences
sentence_1 = "Aam Aadmi Party (AAP) MLA and Delhi minister Atishi was picked by the party."
sentence_2 = "India’s first woman CM and the state’s fourth, Kripalani was 55 years old."

# Encode inputs
inputs = tokenizer(sentence_1, sentence_2, return_tensors='pt')

# Get predictions
outputs = model(**inputs)
logits = outputs.logits

# Check the probabilities of next sentence being true or false
probs = torch.softmax(logits, dim=1)
print(f"Probabilities: {probs}")
```

##3. Abstractive Summarization using BART or T5


```python
!pip install transformers
```

```python
from transformers import BartTokenizer, BartForConditionalGeneration

# Load pre-trained BART model and tokenizer
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large-cnn')
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large-cnn')

# Tokenize the input text
inputs = tokenizer(dataset, max_length=1024, return_tensors='pt', truncation=True)

# Generate summary
summary_ids = model.generate(inputs['input_ids'], num_beams=4, max_length=150, early_stopping=True)
bart_summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
print(f"Abstractive Summary: {bart_summary}")
```

```python
from rouge_score import rouge_scorer

# Define reference and generated summaries
reference_summary = "Your reference summary here"  # Ground truth
generated_summary = bart_summary  # Using the generated summary from BART or T5

# Initialize ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2'], use_stemmer=True)

# Compute ROUGE scores
scores = scorer.score(reference_summary, generated_summary)
print(f"ROUGE-1: {scores['rouge1'].fmeasure}")
print(f"ROUGE-2: {scores['rouge2'].fmeasure}")
```
